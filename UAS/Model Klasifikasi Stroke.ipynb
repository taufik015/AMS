```python
import pandas as pd
```

```python
stroke = pd.read_csv('stroke.csv')
```

```python
data = stroke.copy()
```

```python
target = 'stroke'
```

```python
encode = ['Residence_type', 'gender']
```

```python
data
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy],axis=1) 
    del data[col]
```

```python
data
```

```python
target_mapper = {'No':0, 'yes':1}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['stroke'] = data['stroke'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('stroke',axis=1)
Y = data['stroke']
```

```python
X
```

```python
Y
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,Y)
```

```python
import pickle
pickle.dump(model, open('modelNBC_stroke.pkl','wb'))
```
